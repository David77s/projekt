```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df_car = pd.read_csv('../pliki/car_prices.csv')
```

```python
df_car.info()
```

Ilość ogłoszeń ze względu na rodzaj paliwa

```python
_ = sns.catplot(df_car, x='fuel', kind='count')
```

Ilość ogłoszeń w zależności od ceny samochodów

```python
_ = sns.displot(df_car, x = 'price', bins = 80)
```

Ilość ogłoszeń w zależności od pojemności silnika

```python
_ = sns.displot(df_car, x = 'vol_engine', bins = 80)
```

Ilość ogłoszeń w zależności od roku produkcji danego samochodu

```python
_ = sns.displot(df_car, x = 'year', bins = 80)
```

Ilość ogłoszeń ze względu na markę samochodu

```python
_ = sns.catplot(df_car, y = 'mark', kind='count')
```

Ilość ogłoszeń ze względu na region geograficzny

```python
_ = sns.catplot(df_car, y = 'province', kind='count')
```
